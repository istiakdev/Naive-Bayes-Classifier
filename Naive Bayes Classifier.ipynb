```
import numpy as np 
from google.colab import files
import io
import pandas as pd
```

```
uploaded = files.upload()
```

```
df = pd.read_csv(io.BytesIO(uploaded['Data_csv.csv'] ), sep="," ,  header = 0, squeeze=True)
print(df)
```

```
from sklearn.model_selection import train_test_split
x_train, x_test = train_test_split(df,test_size=0.4, random_state=42)
print(x_train)
```

```
training_set = x_train.to_numpy()
testing_set = x_test.to_numpy()
print(testing_set)
```

```
count1 = sum(map(lambda x : x[2] == 1, training_set))
print(count1)
count2=sum(map(lambda x : x[2] == 2, training_set))
print(count2)
```

```
#class separation
k = 0
j=0
#w1[0]=feature1 of class 1
#w1[1]=feature2 of class 1
#w2[0]=feture1 of class 2
#w2[1]=feature2 of class 2
w1=np.zeros((2,count1))
w2=np.zeros((2,count2))
for i in range (len(training_set)):
  if  training_set[i][2] ==1 :
    w1[0][k]=training_set[i][0]
    w1[1][k]= training_set[i][1]
    k=k+1

  else :
    w2[0][j]=training_set[i][0]
    w2[1][j]= training_set[i][1]
    j=j+1

print (w1)  
print (w2)  
```

```
#mean 
w1m=np.mean(w1, axis = 1)  #class1
w2m=np.mean(w2, axis = 1)  #class2
print(w1m)
print(w2m)
```

```
#covariance
cov1=np.cov(w1)
cov2=np.cov(w2)
print(cov1)
print(cov2)
```

```
#probability
prob1=count1/len(training_set)
print(prob1)
prob2=count2/len(training_set)
print(prob2)
```

```
#classification of testing_set
import math
invCov1 = np.linalg.inv(cov1)
invCov2=np.linalg.inv(cov2)
list1x=[]
list1y=[]
list2x=[]
list2y=[]
count=0
for i in range (len(testing_set)):
  arr=np.array([testing_set[i][0],testing_set[i][1]])
  res=arr-w1m 
  resTmulres=(res[0]*res[0]) + (res[1]*res[1])
  exptemp=-0.5*resTmulres*np.mean(invCov1)  
  expans=math.exp(exptemp)
  g1=1/(math.sqrt((2*3.1416)*(2*3.1416)*np.linalg.det(cov1)) )
  g1a=g1*expans
  g1f= g1a*prob1 



  res2=arr-w2m 
  resTmulres2=(res2[0]*res2[0]) + (res2[1]*res2[1])
  exptemp2=-0.5*resTmulres2*np.mean(invCov2)
  expans2=math.exp(exptemp2)
  g2=1/(math.sqrt((2*3.1416)*(2*3.1416)*np.linalg.det(cov2)) )
  g2a=g2*expans2
  g2f= g2a*prob2 
  if g1f>g2f : 
        
    list1x.append(testing_set[i][0])
    list1y.append(testing_set[i][1])
    if testing_set[i][2]==1:
      count=count+1
  elif g2f>g1f: 
    list2x.append(testing_set[i][0])
    list2y.append(testing_set[i][1])
    if testing_set[i][2]==2:
      count=count+1
```

```
#print(list1x)
print('accuracy',count/len(testing_set)*100,'%') 
```
